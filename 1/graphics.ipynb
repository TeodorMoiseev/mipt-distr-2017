```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Для начала проведем измерения при $P = 4$.

```python
N = np.arange(10000, 110000, 10000)
T_4 = np.array([0.30356, 0.62909, 0.99744, 1.28837, 1.60679, 1.90233, 2.22295, 2.60710, 2.87100, 3.15300])
T_1 = np.array([0.55228, 1.17195, 1.72172, 2.34601, 2.82244, 3.42653, 3.92628, 4.56512, 5.04540, 5.70371])
S = T_1 / T_4
E = S / 4
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('N')
plt.title('P = 4')
plt.grid(ls=':')

plt.plot(N, T_4, '.-', label = 'T')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('N')
plt.title('P = 4')
plt.grid(ls=':')

plt.plot(N, S, '.-', label = 'S')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('N')
plt.title('P = 4')
plt.grid(ls=':')

plt.plot(N, E, '.-', label = 'E')
plt.legend()
plt.show()
```

А теперь проведем измерения при $N = 10000$ при различных $P$.

```python
P = np.array([1, 2, 4, 8, 16])
T = np.array([0.56536, 0.39799, 0.31338, 0.31746, 0.30502])
S = np.full(T.size, T[0]) / T
E = S / P
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P')
plt.title('N = 10000')
plt.grid(ls=':')

plt.plot(P, T, '.-', label = 'T')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P')
plt.title('N = 10000')
plt.grid(ls=':')

plt.plot(P, S, '.-', label = 'S')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('P')
plt.title('N = 10000')
plt.grid(ls=':')

plt.plot(P, E, '.-', label = 'E')
plt.legend()
plt.show()
```
