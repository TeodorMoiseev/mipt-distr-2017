```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Делаем запуски при разном числе потоков при $N = 10000000$ и $m = 10000$. Также запускаем qsort.

```python
qsort_time = 1.98382
P = np.array([1, 2, 4, 8, 16])
T = np.array([2.23709, 1.17162, 1.17061, 1.14947, 1.16390])
S = T[0] / T
E = S / P
```

```python
plt.figure(figsize=(15, 9))
plt.grid(ls=':')
plt.title('T(P)')
plt.xlabel('P')
plt.ylabel('T')

plt.plot(P, T, 'bo-', label='parallel merge sort')
plt.axhline(qsort_time, label='qsort', color='red')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.grid(ls=':')
plt.title('S(P)')
plt.xlabel('P')
plt.ylabel('S')

plt.plot(P, S, 'bo-', label='parallel merge sort')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.grid(ls=':')
plt.title('E(P)')
plt.xlabel('P')
plt.ylabel('E')

plt.plot(P, E, 'bo-', label='parallel merge sort')
plt.legend()
plt.show()
```
